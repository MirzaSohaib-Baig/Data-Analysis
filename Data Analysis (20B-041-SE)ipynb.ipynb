# Clean Data

```python
col_with_no_missing = [col for col in pm.columns if pm[col].isnull().sum() == 0]
col_with_no_missing
```

```python
col_with_missing = [col for col in pm.columns if pm[col].isnull().sum() > 0]
col_with_missing
```

```python
df = pm.dropna(axis = 0)
df.isnull().sum()
df
```

```python
df.columns
```

```python
country = df.groupby('MeetCountry').size().reset_index().rename(columns = {0:'Counts'}).sort_values(by='Counts',ascending=False)
country.head()
```

```python
plt.pie(x = 'Counts',labels = 'MeetCountry',data = country,autopct = '%.2f')
plt.show()
```

```python
col_with_no_missing2 = [col for col in pl.columns if pl[col].isnull().sum() == 0]
col_with_no_missing2
```

```python
col_with_missing2 = [col for col in pl.columns if pl[col].isnull().sum() > 0]
col_with_missing2
```

```python
df2 = pl.dropna(axis = 0)
df2.isnull().sum()
df2
```

```python
bench = df2.groupby('BestBenchKg').size().reset_index().rename(columns = {0:'Counts'}).sort_values(by='Counts',ascending=False).reset_index()
bench.drop('index',axis = 1,inplace = True)
bench.head()
```

```python
sns.barplot(x = 'BestBenchKg',y='Counts',data = bench)
plt.show()
```
